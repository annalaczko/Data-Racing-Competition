```python
def datagen(df, dummy_cols, startnap, endnap):

    df2 = df[ df['CONTRACT_DATE_OF_LOAN_AGREEMENT']>=startnap].copy()
    df = df2[ df2['CONTRACT_DATE_OF_LOAN_AGREEMENT']<endnap].copy()

    cust_df = pd.DataFrame({'BORROWER_ID':df['BORROWER_ID'].unique()})
# TARGET VALTOZO KESZITES
    feltetel=((df['TARGET_EVENT_DAY']>endnap) & (df['TARGET_EVENT_DAY']<endnap+2*365) & (df['TARGET_EVENT']=='K')  )
    df['target']=0
    df.loc[feltetel,'target']=1
    
    stat = df.groupby("BORROWER_ID",as_index=False).agg({'target':'max'})
    cust_df=cust_df.merge(stat,on='BORROWER_ID',how='left')
    cust_df['target']=cust_df['target'].fillna(0)  

    custom_aggregations = {col: 'sum' for col in dummy_cols}
    

    stat = df.groupby("BORROWER_ID",as_index=False).agg(
    {   'CONTRACT_ID':'count',
        'CONTRACT_CREDIT_LOSS': 'sum',
        'CONTRACT_DEPT_SERVICE_TO_INCOME': 'mean',
        'CONTRACT_INCOME': 'sum',
        'CONTRACT_INSTALMENT_AMOUNT': 'mean',
        'CONTRACT_INTEREST_RATE': 'mean',
        'CONTRACT_LGD': 'sum',
        'CONTRACT_LOAN_AMOUNT': 'sum',
        'CONTRACT_LOAN_TO_VALUE_RATIO': 'mean',
        'CONTRACT_MARKET_VALUE': 'sum',
        'CONTRACT_MORTGAGE_LENDING_VALUE': 'sum',
         **custom_aggregations
    }
    )

    columns=[   'BORROWER_ID',
                'CONTRACT_ID_SUM',
                'CONTRACT_CREDIT_LOSS_SUM',
                'CONTRACT_DEPT_SERVICE_TO_INCOME_MEAN',
                'CONTRACT_INCOME_SUM',
                'CONTRACT_INSTALMENT_AMOUNT_MEAN',
                'CONTRACT_INTEREST_RATE_MEAN',
                'CONTRACT_LGD_SUM',
                'CONTRACT_LOAN_AMOUNT_SUM',
                'CONTRACT_LOAN_TO_VALUE_RATIO_MEAN',
                'CONTRACT_MARKET_VALUE_SUM',
                'CONTRACT_MORTGAGE_LENDING_VALUE_SUM'
            ]

    columns.extend(dummy_cols)
    stat.columns=columns
    cust_df=cust_df.merge(stat,on='BORROWER_ID',how='left')

    return cust_df
```

```python
def datagentarget(df, dummy_cols, startnap, endnap):

    df2 = df[ df['CONTRACT_DATE_OF_LOAN_AGREEMENT']>=startnap].copy()
    df3 = df2[df2['TARGET_EVENT_DAY']>=startnap].copy()
    df4 = df3[df3['TARGET_EVENT_DAY']<endnap].copy()
    df = df4[ df4['CONTRACT_DATE_OF_LOAN_AGREEMENT']<endnap].copy()

    cust_df = pd.DataFrame({'BORROWER_ID':df['BORROWER_ID'].unique()})
# TARGET VALTOZO KESZITES
    # feltetel=((df['TARGET_EVENT_DAY']>endnap) & (df['TARGET_EVENT_DAY']<endnap+2*365) & (df['TARGET_EVENT']=='K')  )
    df['target']=1
    # df.loc[feltetel,'target']=1
    
    stat = df.groupby("BORROWER_ID",as_index=False).agg({'target':'max'})
    cust_df=cust_df.merge(stat,on='BORROWER_ID',how='left')
    cust_df['target']=cust_df['target'].fillna(0)  

    custom_aggregations = {col: 'sum' for col in dummy_cols}
    

    stat = df.groupby("BORROWER_ID",as_index=False).agg(
    {   'CONTRACT_ID':'count',
        'CONTRACT_CREDIT_LOSS': 'sum',
        'CONTRACT_DEPT_SERVICE_TO_INCOME': 'mean',
        'CONTRACT_INCOME': 'sum',
        'CONTRACT_INSTALMENT_AMOUNT': 'mean',
        'CONTRACT_INTEREST_RATE': 'mean',
        'CONTRACT_LGD': 'sum',
        'CONTRACT_LOAN_AMOUNT': 'sum',
        'CONTRACT_LOAN_TO_VALUE_RATIO': 'mean',
        'CONTRACT_MARKET_VALUE': 'sum',
        'CONTRACT_MORTGAGE_LENDING_VALUE': 'sum',
         **custom_aggregations
    }
    )

    columns=[   'BORROWER_ID',
                'CONTRACT_ID_SUM',
                'CONTRACT_CREDIT_LOSS_SUM',
                'CONTRACT_DEPT_SERVICE_TO_INCOME_MEAN',
                'CONTRACT_INCOME_SUM',
                'CONTRACT_INSTALMENT_AMOUNT_MEAN',
                'CONTRACT_INTEREST_RATE_MEAN',
                'CONTRACT_LGD_SUM',
                'CONTRACT_LOAN_AMOUNT_SUM',
                'CONTRACT_LOAN_TO_VALUE_RATIO_MEAN',
                'CONTRACT_MARKET_VALUE_SUM',
                'CONTRACT_MORTGAGE_LENDING_VALUE_SUM'
            ]

    columns.extend(dummy_cols)
    stat.columns=columns
    cust_df=cust_df.merge(stat,on='BORROWER_ID',how='left')

    return cust_df
```
