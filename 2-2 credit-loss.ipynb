```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import csv
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import SGDClassifier
```

```python
readIn = pd.read_csv("data/df_dummy_prepped.csv")
df = pd.DataFrame(readIn)
```

```python
input=list(df.columns)
```

```python
input
```

```python
to_remove=['CONTRACT_ID', 'BORROWER_ID', 'CONTRACT_CREDIT_LOSS','CONTRACT_DATE_OF_LOAN_AGREEMENT','CONTRACT_MATURITY_DATE', 'TARGET_EVENT_DAY', 'TARGET_EVENT']
```

```python
new_input = [x for x in input if x not in to_remove]
```

```python
# dummy_cols = []

# with open('data/dummy_cols.csv', 'r') as file:
#     reader = csv.reader(file)
#     for row in reader:
#         dummy_cols=row
```

```python
target= 'CONTRACT_CREDIT_LOSS'
```

```python
X = df[new_input]
y = df[target]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
```

```python
model = GradientBoostingRegressor(n_estimators=100, max_depth=10, random_state=22, warm_start= True)
```

```python
batch_size=1000
```

```python
for i in range(0, len(X_train), batch_size):

    print(f'Processing {i}', end='\r')
    X_batch = X_train[i:i+batch_size]
    y_batch = y_train[i:i+batch_size]

    # Modell tanítása a batch-en
    model.fit(X_batch, y_batch)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```
